# Clustering on vehicle dataset

To determine the group in which a particular vehicle lies.

```
# Importing the required modules
import numpy as np 
import pandas as pd
from scipy import ndimage 
from scipy.cluster import hierarchy 
from scipy.spatial import distance_matrix 
from matplotlib import pyplot as plt 
from sklearn import manifold, datasets 
from sklearn.cluster import AgglomerativeClustering 
from sklearn.datasets.samples_generator import make_blobs 
%matplotlib inline
```

```
# Reading the data 
df = pd.read_csv('cars_clus.csv')

df.head()
```

```
# Cleaning the data by driping the rows which have null value

df[[ 'sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']] = df[['sales', 'resale', 'type', 'price', 'engine_s',
       'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap',
       'mpg', 'lnsales']].apply(pd.to_numeric, errors='coerce') # Changing null to NaN

df = df.dropna()
df = df.reset_index(drop = True)
```

```
df.head() # data after cleaning
```

```
# Selecting the required feature set 
features = df[['engine_s',  'horsepow', 'wheelbas', 'width', 'length', 'curb_wgt', 'fuel_cap', 'mpg']]
```

```
# Normalizing our feature set using MinMaxScaler
from sklearn.preprocessing import MinMaxScaler
x = features.values
feature_mat = MinMaxScaler().fit_transform(x)
feature_mat[0:5]
```

Clustering using scipy

```
import scipy

# Calculating the distance matrix

length = features.shape[0]
dist_mat = scipy.zeros([length,length])
for i in range(length):
  for j in range(length):
    dist_mat[i,j] = scipy.spatial.distance.euclidean(feature_mat[i], feature_mat[j])
```

```
dist_mat.shape
```

```
import scipy.cluster.hierarchy
Z = hierarchy.linkage(dist_mat, 'complete')
```

```
from scipy.cluster.hierarchy import fcluster
depth = 3 
cluster = fcluster(Z , depth, criterion = 'distance')
```

```
cluster
```

```
depth = 4 
clusters = fcluster(Z ,depth, criterion= 'maxclust')
clusters
```

Plotting the dendrograms for generated clusters

```
def llf(id):
  return '[%s %s %s]' % (df['manufact'][id],df['model'][id],int(float(df['type'][id])))

fig= plt.figure(figsize=(15,30))

dendro = hierarchy.dendrogram(Z ,leaf_label_func = llf , leaf_font_size = 12, orientation = 'right', leaf_rotation=0)
```

##Clustering using sklearn module

```
dist_matrix = distance_matrix(feature_mat, feature_mat)
print(dist_matrix)
```

```
agglom = AgglomerativeClustering(n_clusters = 6 , linkage = 'complete')
agglom.fit(feature_mat)
agglom.labels_
```

```
# adding clusters to the dataFrame
df['cluster'] = agglom.labels_
```

```
df.head()
```

```
# Visulaizing the clusters 
import matplotlib.cm as cm

n_clusters = max(agglom.labels_) + 1

colors = cm.rainbow (np.linspace(0,1,n_clusters))
cluster_labels = list(range(n_clusters))

plt.figure(figsize=(16,14))

for color, label in zip(colors, cluster_labels):
  subset = df[df.cluster == label]
  for i in subset.index: 
    plt.text(subset.horsepow[i], subset.mpg[i], str(subset['model'][i]), rotation = 25)

  plt.scatter(subset.horsepow, subset.mpg, s= subset.price*10, c=color, label='cluster'+str(label),alpha=0.5)


plt.legend()
plt.title("Clusters")
plt.xlabel("Horsepower")
plt.ylabel("mpg")
plt.show()
```

```
df.groupby(['cluster','type'])['cluster'].count()
# type 0 refers to trucks and type 1 refers to cars
```

```
agg_cars = df.groupby(['cluster','type'])['horsepow','engine_s','mpg','price'].mean()
agg_cars
```

```
#potting the clusters based on their type
plt.figure(figsize=(16,10))
for color, label in zip(colors, cluster_labels):
    subset = agg_cars.loc[(label,),]
    for i in subset.index:
        plt.text(subset.loc[i][0]+5, subset.loc[i][2], 'type='+str(int(i)) + ', price='+str(int(subset.loc[i][3]))+'k')
    plt.scatter(subset.horsepow, subset.mpg, s=subset.price*20, c=color, label='cluster'+str(label))
plt.legend()
plt.title('Clusters')
plt.xlabel('horsepow')
plt.ylabel('mpg')
```
